```python
import numpy as np
import umap
import plotly.express as px
import pandas as pd
```

```python
# Load the new "_full" artifacts
embeddings = np.load('processed/embeddings_full.npy')
labels = np.load('processed/labels_full.npy')
class_mapping = np.load('processed/class_mapping.npy', allow_pickle=True).item()

print(f"Loaded full embeddings shape: {embeddings.shape}")
```

```python
print("Running UMAP... This might take a few seconds.")

# Initialize UMAP to squash 2048 dimensions down to 3
reducer = umap.UMAP(n_components=3, random_state=42, n_neighbors=15, min_dist=0.1)

# Fit and transform the data
embeddings_3d = reducer.fit_transform(embeddings)

print(f"UMAP complete! New shape: {embeddings_3d.shape}")
```

```python
# Map the numerical labels back to their actual string names (e.g., "ADONIS")
species_names = [class_mapping[label] for label in labels]

# Create a Pandas DataFrame to easily feed into Plotly
df = pd.DataFrame({
    'x': embeddings_3d[:, 0],
    'y': embeddings_3d[:, 1],
    'z': embeddings_3d[:, 2],
    'Species': species_names
})

# Generate the interactive 3D scatter plot
fig = px.scatter_3d(
    df, x='x', y='y', z='z',
    color='Species',
    title='The Visual Specimen Vault: 3D Latent Space',
    opacity=0.8,
    hover_name='Species'
)

# Adjust marker size for better visibility
fig.update_traces(marker=dict(size=5))

# Display the plot right here in VS Code
fig.show()
```

```python
# Save the full 3D coordinates
np.save('processed/embeddings_3d_full.npy', embeddings_3d)
print("Full 3D coordinates saved successfully!")
```
